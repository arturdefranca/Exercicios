```python
#Importação de bibliotecas
import csv #Manipula arquivos csv

#Função para centralizar textos entre bordas
#                         OBSERVAÇÕES
#  Para obter o resultado da função abaixo utilizei os conceitos abaixo:
#
# 1)  Para personalizar o tipo de borda, pesquise tabela ASCII e encontre um tipo para substituir o '│', geralmente
#   em windows segure a tecla <ALT> e digite uma sequência numérica e solte. Em meu caso usei o <ALT> + 179 conforme coluna
#   decimal do arquivo http://ic.unicamp.br/~everton/aulas/hardware/tabelaASCII.pdf
# 2)  O método utilizado para centralizar foi o center()
#   modo de uso: str.center(n,[x]) - onde n = tamanho e x = caracter opcional
#   esse método permite escolher um texto (str) para aparecer no meio de uma expressão de tamanho (n) cercado por (x) caracter
#   p.ex.: 'Oi'.center(10,'-') ==> ----Oi----
#   Vale a pena ver: ljust() e rjust()
# 3)  Visando alterar o padrão do print() que imprime 'enter' no fim do texto usa-se 'end=', no caso o caracter especial '\r'
#   o cursor do texto permanecerá na mesma linha de onde parou o texto exibido.
# 4)  Função com argumento padrão, ao utilizar 'paragrafo=True' diz-se ao python que tal argumento é padrão, ou seja,
#   caso não conter na solicitação considere esse valor como válido, não necessariamente tem de ser booleano.


def centraliza(texto, tamanho, paragrafo=True):
    if paragrafo:
        print('│',end='')
        print(texto.center(tamanho," "), end='')
        print('│')
    else:
        print('│',end='\r')
        print(texto.center(tamanho," "), end='\r')
        print('│',end='\r')

        
# Preparação de variáveis globais

continua_fluxo = True
controle = 0

while continua_fluxo:
    #Preparação de variáveis locais dentro do loop while
    frq_primeiro_nome = 0
    frq_nome_completo = 0
    frq_masculino_primeiro_nome = 0
    frq_masculino_nome_completo = 0
    frq_feminino_primeiro_nome = 0
    frq_feminino_nome_completo = 0
    porc_feminino_nome_completo = 0
    porc_feminino_primeiro_nome = 0
    porc_masculino_nome_completo = 0
    porc_masculino_primeiro_nome = 0
    nome_encontrado = False

    # Solicitar que usuário digite um nome
    print()
    nome_escolhido = input('Digite um NOME para ser analisado ou "SAIR" para finalizar: ')

    # Verificar se é para sair do sistema ou continuar
    if nome_escolhido.upper() == 'SAIR':
        print('┌────────────────────────────────────────────────────────────────────────────────────────────────────┐')
        centraliza('Fechando sistema de análise de nomes',100)
        print('└────────────────────────────────────────────────────────────────────────────────────────────────────┘')
        continua_fluxo = False
        break
    
    #nome_escolhido = nome_escolhido + ' ' #adicionar espaço para diferenciar de nomes que contém outro p.ex.: Maria não quero Mariano
    
    # Abrir arquivo que contém nomes e sexo de candidatos à cargos políticos desde 1998 - 2016
    arquivo = open('nomecandidatos.csv', encoding='utf8')

    #Iniciando Interface com usuário
    print('┌────────────────────────────────────────────────────────────────────────────────────────────────────┐')
    print('├───────────────────────────────────────── A N Á L I S E ────────────────────────────────────────────┤')
    centraliza(' ',100)

    #
    # Percorrendo os registros, transformados em dicionário, do arquivo csv aberto no passo anterior
    # Manuasear os seguintes campos do arquivo
    # - nomecand     --> Nome completo do candidato
    # - primnomecand --> Primeiro nome do candidato
    # - sexo         --> Sexo do candidato
    #
    
    for registro in csv.DictReader(arquivo):
        if (nome_escolhido.lower().capitalize() + ' ') in registro['nomecand']:
            frq_nome_completo += 1
            centraliza(registro['nomecand'] + ' / '+ registro['sexo'],100,False)

            if registro['sexo'] == 'Feminino':
                frq_feminino_nome_completo += 1
            elif registro['sexo'] == 'Masculino':
                frq_masculino_nome_completo += 1

            if (nome_escolhido.lower().capitalize()) == registro['primnomecand']:
                frq_primeiro_nome += 1
                if registro['sexo'] == 'Feminino':
                    frq_feminino_primeiro_nome += 1
                elif registro['sexo'] == 'Masculino':
                    frq_masculino_primeiro_nome +=1

        if frq_nome_completo > 0:
            nome_encontrado = True

            porc_feminino_nome_completo = round((frq_feminino_nome_completo / frq_nome_completo)* 100, 2)
            porc_masculino_nome_completo = round((frq_masculino_nome_completo / frq_nome_completo)* 100, 2)

        if frq_primeiro_nome == 0:
            sexo_nome = 'que não se determina o sexo'
        else:
            if porc_feminino_nome_completo > 50:
                sexo_nome = 'mulher'
            else:
                sexo_nome = 'homem'

    if nome_encontrado == False:
        print('┌────────────────────────────────────────────────────────────────────────────────────────────────────┐')
        centraliza('O nome que você digitou não foi encontrado',100)
        print('└────────────────────────────────────────────────────────────────────────────────────────────────────┘')
        continua_fluxo = False
        break


    centraliza(f'{nome_escolhido.upper()} apareceu {frq_nome_completo} vezes,',100)
    centraliza(f'incluindo {frq_primeiro_nome} vezes que era o primeiro nome.',100)
    centraliza(' ',100)

    print('├═══════════════════════════════════════ R E S U L T A D O ══════════════════════════════════════════┤')
    centraliza(' ',100)

    if sexo_nome == 'mulher':
        centraliza(f'Baseando-se apenas nesse cadastro, {porc_feminino_nome_completo} porcento se refere a mulheres,',100)
    elif sexo_nome == 'homem':
        centraliza(f'Baseando-se apenas nesse cadastro, {porc_masculino_nome_completo} porcento se refere a homens,',100)
    elif sexo_nome == 'que não se determina o sexo':
        centraliza(f'Não foi possível detectar sexo por esse nome, provavelmente é um sobrenome.',100)

    centraliza(f'ou seja, podemos dizer que {nome_escolhido.upper()} é um nome de {sexo_nome}.',100)
    centraliza(' ',100)
    print('└────────────────────────────────────────────────────────────────────────────────────────────────────┘')
    
    controle += 1
    if controle > 5:
        break
    arquivo.close()
    
nome_encontrado = False
```
